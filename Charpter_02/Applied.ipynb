## Exercise 2.3:

3. We now revisit the bias-variance decomposition.

    (a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

    (b) Explain why each of the five curves has the shape displayed in part (a).

### Explanation

(a)


```python
import matplotlib.pyplot as plt
import numpy as np

# Create a range for the x-axis representing flexibility
x = np.arange(0, 10, 0.5)

# Define the curves
# These are illustrative functions to mimic the general shapes of the curves
bias2 = (10 - x)**2 / 50  # Squared bias
variance = x**2 / 50  # Variance
training_error = 2 - 0.8*x + 0.2*x**2 - 0.02*x**3 + 0.0005 * x ** 4# Training error
test_error = bias2 + variance + 1  # Test error
bayes_error = np.full_like(x, 1)  # Bayes (irreducible) error

# Plot the curves
plt.figure(figsize=(10, 6))
plt.plot(x, bias2, label='Squared Bias', color='blue', linestyle='--')
plt.plot(x, variance, label='Variance', color='red', linestyle='-.')
plt.plot(x, training_error, label='Training Error', color='green', linestyle=':')
plt.plot(x, test_error, label='Test Error', color='purple', linestyle='-')
plt.plot(x, bayes_error, label='Bayes Error', color='orange', linestyle='-')

# Add labels and title
plt.xlabel('Flexibility')
plt.ylabel('Error')
plt.title('Bias-Variance Tradeoff and Error Curves')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

(b) (b) Explanation of Curve Shapes

1. Squared Bias: Initially decreases because more flexible models can capture complex patterns, reducing bias. However, it might flatten out or increase slightly as the model becomes too complex and starts overfitting.

2. Variance: Increases with flexibility because more flexible models are capable of fitting noise in the training data, leading to greater variability in predictions.

3. Training Error: Decreases with flexibility because more flexible models can better fit the training data, resulting in lower training error.

4. Test Error: Initially decreases due to reduced bias and improved fit to training data. However, it starts to increase as the model becomes overly complex and starts to overfit, leading to higher test error.

5. Bayes (Irreducible) Error: Remains constant because it represents the inherent noise or irreducible error in the data that cannot be eliminated by any model.

## Exercise 2.6:

6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a non-parametric approach)? What are its disadvantages?

### Explanation

Parametric Approach: 
    A parametric approach to statistical learning involves assuming a specific form for the function that relates the predictors to the response. The model is characterized by a set of parameters, and learning involves estimating these parameters from the data.

Non-Parametric Approach:
    A non-parametric approach does not assume a specific form for the function relating the predictors to the response. Instead, it aims to learn the function directly from the data, allowing for more flexibility in capturing complex patterns.

Advantages of Parametric Approaches:

1. Simplicity and Interpretability

2. Low Variance

3. Faster Computation

4. Avoid Overfitting

Disadvantages of Parametric Approaches:

1. High bias (complex and non-linear dataset)

2. Inflexibility
