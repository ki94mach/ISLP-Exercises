### (a)

```python
import pandas as pd
college = pd.read_csv("data/College.csv")
college.head()
```

### (b)

```python
college2 = pd.read_csv("data/College.csv", index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'},axis=1) 
college3 = college3.set_index('College')
college = college3
```

### (c)

```python
college.describe()
```

### (d)

```python
c = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[c])
```

### (e)

```python
college.boxplot('Outstate', 'Private')
```

### (f)

```python
college['Elite'] = pd.cut(college['Top10perc'],
                          [0, 50, 100],
                          labels=['No', 'Yes'])
print(college.Elite.value_counts())
college.boxplot('Outstate', 'Elite')
```

### (g)

```python
from matplotlib.pyplot import subplots
fig, axes = subplots(2, 2, figsize=(12, 12))
college.hist('Apps', ax=axes[0, 1], bins=50, color="purple")
college.hist('Enroll', ax=axes[0, 0], bins=30,color="pink")
college.hist('Top10perc', ax=axes[1, 0], bins=30, color="red")
college.hist('Top25perc', ax=axes[1, 1], bins=30, color="green")
```
